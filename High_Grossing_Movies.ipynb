```python
%matplotlib inline
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import math
```

```python
link1 = "http://www.the-numbers.com/movie/records/All-Time-Domestic-Box-Office"
r = requests.get(link1)
soup = BeautifulSoup(r.text, "lxml")
```

```python
rawdata = {'Rank':[], 'Release_year':[], 'Movie_title':[], 'Domestic_boxoffice':[], 'Inter_boxoffice':[], 'World_boxoffice':[]}
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF8')
allrows = soup.find_all("tr")
for row in allrows [1: 100 +1]:
    el = row.find_all('td')
    Rank = el[0].contents[0]
    Release_year = el[1].contents[-1]
    Movie_title = el[2].contents[-1].find('a').contents[-1]
    Domestic_boxoffice = el[3].contents[-1]
    Inter_boxoffice = el[4].contents[-1]
    World_boxoffice = el[5].contents[-1]
    try:
        rRank = int(Rank.strip())
    except:
        print("bad data for Rank", Rank)
    try:
        rRelease_year = int(Release_year.strip())
    except:
        print("bad data for Release_year", Release_year)
    try:
        mMovie_title = Movie_title.strip()
    except:
        print("bad data for Movie_title", Movie_title)
    try:
        dDomestic_boxoffice = Domestic_boxoffice.strip()
        dbDomestic_boxoffice = dDomestic_boxoffice[1:]
        dbbDomestic_boxoffice=int(locale.atoi(dbDomestic_boxoffice))
    except:
        print("bad data for Domestic_boxoffice", Domestic_boxoffice)
    try:
        iInter_boxoffice = Inter_boxoffice.strip()
        iiInter_boxoffice = iInter_boxoffice[1:]
        iiiInter_boxoffice = int(locale.atoi(iiInter_boxoffice))
    except:
        print("bad data for Inter_boxoffice", Inter_boxoffice)
    try:
        wWorld_boxoffice = World_boxoffice.strip()
        wwWorld_boxoffice = wWorld_boxoffice[1:]
        wwwWorld_boxoffice = int(locale.atoi(wwWorld_boxoffice))
    except:
        print("bad data for World_boxoffice", World_boxoffice)
    rawdata['Rank'].append(rRank)
    rawdata['Release_year'].append(rRelease_year)
    rawdata['Movie_title'].append(mMovie_title)
    rawdata['Domestic_boxoffice'].append(dbbDomestic_boxoffice)
    rawdata['Inter_boxoffice'].append(iiiInter_boxoffice)
    rawdata['World_boxoffice'].append(wwwWorld_boxoffice)
```

```python
rawdata
```

```python
df = pd.DataFrame(rawdata,range(1,101))
#df.head(10)
df
```

```python
df.to_csv('grossboxoffice.csv')
```

```python

def get_adjusted_price(year, price):
    req_url="https://www.statbureau.org/calculate-inflation-price-json?country=united-states&start={}/1/1&end=2016/8/1&amount={}"
    resp = requests.get(req_url.format(year,price))
    if resp.status_code ==200:
        #adj_price = float(resp.json().replace('$','').replace(',',''))
        adj_price = float(resp.json().strip('$').strip(',').replace(' ', ''))
    else:
        adj_price = math.nan
    return adj_price
print(get_adjusted_price(1999, 1.00))
```
